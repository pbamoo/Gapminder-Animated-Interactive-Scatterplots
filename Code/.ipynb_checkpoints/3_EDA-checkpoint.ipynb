### Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("white")
import plotly_express as px
import imageio
from ipynb.fs.full.utils import transform_data, missing_data
import warnings
warnings.simplefilter(action = 'ignore', category = FutureWarning)
pd.options.display.max_columns = None
```

```python
# Read files
fert = pd.read_csv("../data/gapminder_total_fertility.csv", index_col=0)
population = pd.read_excel("../data/gapminder_population.xlsx", index_col = 0, engine='openpyxl')
life = pd.read_excel("../data/gapminder_lifeexpectancy.xlsx", index_col=0, engine='openpyxl')
```

```python
df = transform_data(fert, population, life)
df = missing_data(df)
```

```python
#Visualize subset of e.g 3 countries (Canada, Ghana, Germany)
df_subset = df.loc[df['country'].isin(['Canada', 'Ghana', 'Germany'])]
df_subset
```

```python
sns.lineplot(x='year', y='fertility_rate', hue='country',
            data=df_subset)
plt.show()
```

```python
sns.lineplot(x='year', y='total_population', hue='country',
            data=df_subset)
plt.show()
```

```python
sns.lineplot(x='year', y='life_expectancy', hue='country',
            data=df_subset)
plt.show()
```

```python
# We are going to create an animated scatterplot for each yearshowing the correlation between 
# life expectancy and fertility, with the size of the symbols represented by the population
my_dpi=96
df_subset['country']=pd.Categorical(df_subset['country'])
#df_subset = df_subset[ df_subset.year >= 1960 ]
for i in df_subset.year.unique():    
    fig = plt.figure(figsize=(680/my_dpi, 480/my_dpi), dpi=my_dpi)
    tmp=df_subset[ df_subset.year == i ]
    plt.scatter(tmp['life_expectancy'], tmp['fertility_rate'] , s=tmp['total_population']/100000 , 
    c=tmp['country'].cat.codes, cmap="Accent", alpha=0.6, edgecolors="white", linewidth=2)
    #plt.yscale('log')
    plt.xlabel("life_expectancy")
    plt.ylabel("fertility_rate")
    plt.title("Year: "+str(i) )
    plt.axis((0, 100, 0, 10))
    filename='Gapminder'+str(i)+'.png'
    plt.savefig("../Images/Gapminder_" +str(i), dpi=96)
    plt.gca()
```

```python
import imageio

images = []

for i in range(1960, 2015):
    df_subset = '../Images/Gapminder_{}.png'.format(i)
    images.append(imageio.imread(df_subset))

imageio.mimsave('../Images/Gapminder.gif', images, fps=5)
```
