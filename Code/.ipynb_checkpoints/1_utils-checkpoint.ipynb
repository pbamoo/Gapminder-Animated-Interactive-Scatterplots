```python
# Read files
fert = pd.read_csv("../data/gapminder_total_fertility.csv", index_col=0)
population = pd.read_excel("../data/gapminder_population.xlsx", index_col = 0, engine='openpyxl')
life = pd.read_excel("../data/gapminder_lifeexpectancy.xlsx", index_col=0, engine='openpyxl')
```

We can immediately see from the three dataFrames above that there lot of missing data while the shapes aren't the same. The population dataFrame for instance captured data only decade wise from the 1800s to 1950 while the other 2 did so yearly. We are therefore going subset and work with year > 1950, do some preprocessing and merge all three dataFrames.

### Preprocessing the Data
* Transforming Dataframe
* Handling NaNs

```python
# Transform columns with dtype string to integer
fert.columns = fert.columns.astype(int)

# Rename index to 'country'
fert.index.name = 'country'
population.index.name = 'country'
life.index.name = 'country'

# Reset the index
fert = fert.reset_index()
population = population.reset_index()
life = life.reset_index()

# Convert the tables into long format, melt & merge
fert = fert.melt(id_vars='country', var_name='year', value_name='fertility_rate')
population = population.melt(id_vars='country', var_name='year', value_name='total_population')
life = life.melt(id_vars='country', var_name='year', value_name='life_expectancy')

df = fert.merge(population).merge(life)
df = df[ df.year >= 1950 ]
df.reset_index(drop=True, inplace=True)

print('\033[1m' + '\033[94m' + 'Table Schema and Data Types:'.center(80)+ '\033[0m')
print(df.info())
print('\n')
print('\033[1m' + '\033[94m' + 'Descriptive Statistics:'.center(80)+ '\033[0m')
print(df.describe())
print('\n')
missing_values = df.isnull().sum()
missing_per = (missing_values/df.shape[0])*100
missing_table = pd.concat([missing_values,missing_per], axis=1, ignore_index=True) 
missing_table.rename(columns={0:'Total Missing',1:'% Missing'}, inplace=True)
print('\033[1m' + '\033[94m' + 'Missing Data Info:'.center(80) + '\033[0m')
print(missing_table)
print('\n')

print('\033[1m' + '\033[94m' + 'DataFrame after Transformation:'.center(80)+ '\033[0m')
df
```

```python
# Drop the missing values
df = df.dropna()

print('\033[1m' + '\033[94m' + 'Missing Years'.center(80) + '\033[0m')
years = df['year'].unique()
missing_years = [y for y in range(min(years), max(years)+1) if y not in years]
print('missing_years=', missing_years)
print('\n')
missing_values = df.isnull().sum()
missing_per = (missing_values/df.shape[0])*100
missing_table = pd.concat([missing_values,missing_per], axis=1, ignore_index=True) 
missing_table.rename(columns={0:'Total Missing',1:'% Missing'}, inplace=True)
print('\033[1m' + '\033[94m' + 'Missing Data Info:'.center(80) + '\033[0m')
print(missing_table)
print('\n')
print('\033[1m' + '\033[94m' + 'DataFrame after dropping NaNs:'.center(80)+ '\033[0m')
df
```
